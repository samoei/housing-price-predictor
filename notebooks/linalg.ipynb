```python
import matplotlib.pyplot as plt
```

```python
u = np.array([2,5])
v = np.array([3,1])
```

```python
x_cords, ycords = zip(u, v)
plt.scatter(x_cords, ycords, color=["r", "b"])
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
```

```python
def plot_vector2d(vector2d, origin=[0,0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1], head_width=0.2, head_length=0.3, length_includes_head=True, **options)
```

```python
plot_vector2d(u, color="r")
plot_vector2d(v, color="b")
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
```

```python
plt.figure()
plt.plot([0, 1], [0, 1])  # Plot a line

plt.arrow(0.2, 0.2, 0.4, 0.4, color='r', width=0.05)
plt.arrow(0.5, 0.2, -0.2, 0.4, color='g', width=0.02)

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```
